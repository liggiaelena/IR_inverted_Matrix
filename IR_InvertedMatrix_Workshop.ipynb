```python
import os
import requests
import feedparser

# Example: Download blog posts from a public RSS feed and save as .txt files
def fetch_and_save_blog_posts(rss_url, save_folder, max_posts=20):
    feed = feedparser.parse(rss_url)
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    count = 0
    for entry in feed.entries:
        if count >= max_posts:
            break
        title = entry.get('title', 'untitled')
        content = entry.get('summary', '')
        # Clean filename
        filename = f"blog_{count+1}.txt"
        filepath = os.path.join(save_folder, filename)
        with open(filepath, 'w', encoding='utf-8') as f:
            f.write(f"{title}\n\n{content}")
        count += 1
    print(f"Saved {count} blog posts to {save_folder}")

# Example RSS feed: Python Software Foundation blog
rss_url = 'https://pyfound.blogspot.com/feeds/posts/default?alt=rss'
fetch_and_save_blog_posts(rss_url, 'sample_docs', max_posts=20)
```

```python
# Example: Load text files from a folder
import os

def load_documents(folder_path):
    documents = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
                documents.append(file.read())
    return documents

# Replace 'sample_docs/' with your actual folder
documents = load_documents('sample_docs/')
print(f"Loaded {len(documents)} documents.")
```

## ✂️ Step 2: Tokenizer


### 🗣 Instructor Talking Point:
> The tokenizer breaks raw text into a stream of words (tokens). This is the foundation for every later step in IR and NLP.

### 🔧 Your Task:
- Implement a basic tokenizer that splits text into lowercase words.
- Handle punctuation removal and basic non-alphanumeric filtering.


```python
import re

def tokenize(text):
    # Lowercase and keep only alphanumeric token chunks.
    # This strips punctuation and splits text into words/numbers.
    return re.findall(r"[a-z0-9]+", text.lower())

# Test on one document
tokens = tokenize(documents[0])
print(tokens[:20])  # Preview first 20 tokens
```

## 🔁 Step 3: Normalization Pipeline (Stemming, Stop Word Removal, etc.)


### 🗣 Instructor Talking Point:
> Now we normalize tokens: remove stop words, apply stemming or affix stripping. This reduces redundancy and enhances search accuracy.

### 🔧 Your Task:
- Use `nltk` to remove stopwords and apply stemming.


```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords', quiet=True)
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def normalize_tokens(tokens):
    # Keep alphabetic tokens, remove stopwords, then stem.
    normalized = []
    for token in tokens:
        if token.isalpha() and token not in stop_words:
            normalized.append(stemmer.stem(token))
    return normalized

# Example: normalize one document
norm_tokens = normalize_tokens(tokens)
print(norm_tokens[:20])
```

### 🗣 Talking Point: Stemming and Stop Words (with `applic`)

> In normalization, we reduce noise before indexing.

- **Stop words** are very common words (like *the*, *is*, *and*) that usually do not add search meaning.
- We remove them so the index focuses on informative terms.

- **Stemming** reduces related words to a common root.
- For example, **application**, **applications**, and **applying** can all become **`applic`** with Porter Stemmer.

Why this helps:
- Smaller vocabulary and less redundancy in the inverted matrix.
- Better matching between query terms and document terms, even when word forms differ.

In short, stop-word removal removes low-value words, and stemming groups word variants into one searchable concept (`applic`).

## 🔍 Step 4: Inverted Index


### 🗣 Instructor Talking Point:
> We now map each normalized token to the list of document IDs in which it appears. This is the core structure that allows fast Boolean and phrase queries.

### 🔧 Your Task:
- Build the inverted index using a dictionary.
- Add code to support phrase queries using positional indexing.


```python
from collections import defaultdict

def build_inverted_index(documents):
    # term -> set(doc_id) during construction to avoid duplicates
    index = defaultdict(set)
    
    for doc_id, text in enumerate(documents):
        tokens = tokenize(text)
        norm_tokens = normalize_tokens(tokens)
        for term in norm_tokens:
            index[term].add(doc_id)
    
    # Convert sets to sorted lists for stable, readable output
    return {term: sorted(doc_ids) for term, doc_ids in index.items()}

inverted_index = build_inverted_index(documents)
print(dict(list(inverted_index.items())[:10]))  # Preview first 10 terms
```

### 🗣 Talking Point: Understanding the Inverted Matrix with `applic`

> An inverted matrix (inverted index) flips the document view:
- Instead of storing **document -> words**, it stores **word -> documents**.
- This makes search faster because we jump directly to the documents that contain a term.

**Example with `applic`**
- After stemming, words like *application*, *applications*, and *applying* can map to `applic`.
- In the inverted matrix, you may see something like: `applic -> [0, 2, 3, 4, 5, 6, 8, 14, 18]`.
- This means documents 0, 2, 3, 4, 5, 6, 8, 14 and 18 contain that concept.

This structure is the foundation for fast keyword search, Boolean queries, and phrase-query extensions.

## 🧪 Test: Phrase Queries


### 🗣 Instructor Talking Point:
> A phrase query requires the exact sequence of terms (e.g., "machine learning"). To support this, extend the inverted index to store positions, not just docIDs.

### 🔧 Your Task:
- Implement 2 phrase queries **using the inverted matrix**.
- Demonstrate that they return the correct documents.


```python
from collections import defaultdict

# Example phrase queries (you can change these to your own)
query1 = "python software"
query2 = "open source"

def phrase_query(query, documents):
    query_terms = normalize_tokens(tokenize(query))
    if not query_terms:
        return []

    # Build a positional index: term -> doc_id -> [positions]
    positional_index = defaultdict(lambda: defaultdict(list))
    for doc_id, text in enumerate(documents):
        doc_terms = normalize_tokens(tokenize(text))
        for pos, term in enumerate(doc_terms):
            positional_index[term][doc_id].append(pos)

    # Candidate docs must contain all query terms
    term_docs = [set(positional_index[term].keys()) for term in query_terms]
    if not term_docs:
        return []
    candidate_docs = set.intersection(*term_docs)

    results = []
    for doc_id in sorted(candidate_docs):
        first_term_positions = positional_index[query_terms[0]][doc_id]
        other_positions = [
            set(positional_index[term][doc_id]) for term in query_terms[1:]
        ]

        # Check whether terms appear in consecutive positions
        for start_pos in first_term_positions:
            if all((start_pos + offset + 1) in other_positions[offset] for offset in range(len(other_positions))):
                results.append(doc_id)
                break

    return results

# Run phrase queries on the loaded documents
results1 = phrase_query(query1, documents)
results2 = phrase_query(query2, documents)

print(f"Documents containing the phrase '{query1}': {results1}")
print(f"Documents containing the phrase '{query2}': {results2}")
```

### 🗣 Talking Point: Why Phrase Queries Are Strict

> A phrase query checks for exact word sequence and adjacency after normalization.
> It does **not** behave like a loose keyword search.

- Has only **python**: **not returned** (missing second term).
- Has only **software**: **not returned** (missing first term).
- Has both terms but far apart: **not returned** (not consecutive positions).
- Has **python software** as an exact phrase: **returned**.

In short, phrase search requires **both terms, in order, side by side**.

### 🗣 Talking Point: Positional Index vs TF-IDF

**Positional Index -> "Where is this word?"**
- Stores the position of each word inside each document.
- Great for finding consecutive words (bigrams, phrases).
- Example: Is **machine** right before **learning**? ✅

**TF-IDF -> "How important is this word in this document?"**
- **TF**: how many times the word appears in that document.
- **IDF**: how rare the word is across all documents (rarer = more valuable).
- Together, they estimate which document is most relevant for a search term.

**Quick distinction:**
- Positional index = best for exact phrase matching.
- TF-IDF = best for relevance ranking among matching documents.

```python
# Implement Positional Indexes: Store term positions for each document
from collections import defaultdict
import tracemalloc
import sys

def build_positional_index(documents):
    positional_index = defaultdict(lambda: defaultdict(list))
    for doc_id, text in enumerate(documents):
        tokens = normalize_tokens(tokenize(text))
        for pos, token in enumerate(tokens):
            positional_index[token][doc_id].append(pos)
    return positional_index

def deep_getsizeof(obj, seen=None):
    # Approximate recursive size of Python objects in bytes.
    if seen is None:
        seen = set()
    obj_id = id(obj)
    if obj_id in seen:
        return 0
    seen.add(obj_id)

    size = sys.getsizeof(obj)
    if isinstance(obj, dict):
        size += sum(deep_getsizeof(k, seen) + deep_getsizeof(v, seen) for k, v in obj.items())
    elif isinstance(obj, (list, tuple, set, frozenset)):
        size += sum(deep_getsizeof(i, seen) for i in obj)
    return size

# Build positional index for the loaded documents and track memory
tracemalloc.start()
positional_index = build_positional_index(documents)
current_mem, peak_mem = tracemalloc.get_traced_memory()
tracemalloc.stop()

# Memory report
index_size_bytes = deep_getsizeof(positional_index)
print(f"Approx. index size in memory: {index_size_bytes / (1024**2):.2f} MB")
print(f"Peak memory during build: {peak_mem / (1024**2):.2f} MB")

# Example: Show positions for a sample term
sample_term = 'python'  # Change to any term you want to inspect
print(f"Positions for term '{sample_term}':")
for doc_id, positions in positional_index[sample_term].items():
    print(f"  Document {doc_id}: {positions}")
```

### 📊 Comparison: Positional Index vs. Term Document Count Matrix

A **Positional Index** stores not only which documents a term appears in, but also the exact positions of each term within those documents. In contrast, a **Term Document Count Matrix** (TDCM) simply records the frequency of each term in each document, without any positional information.

| Feature                      | Positional Index                | Term Document Count Matrix (TDCM) |
|------------------------------|---------------------------------|-----------------------------------|
| Stores term positions        | Yes                             | No                                |
| Supports phrase/bigram search| Yes                             | No                                |
| Memory usage                 | Higher                          | Lower                             |
| Query speed for phrases      | Fast                            | Slow (requires scanning)          |
| Useful for                   | Phrase queries, proximity search | Keyword frequency analysis         |
| Implementation complexity    | Higher                          | Lower                             |

**Summary:**
- Use a positional index for advanced search features like phrase and proximity queries.
- Use a TDCM for simple keyword frequency analysis and ranking.

```python
# Calculate Inverse Document Frequency (IDF) for each term using the positional index
import math

def compute_idf_from_positional_index(positional_index, total_docs):
    idf_scores = {}
    if total_docs <= 0:
        return idf_scores

    for term, doc_positions in positional_index.items():
        doc_freq = len(doc_positions)  # number of documents containing the term
        if doc_freq > 0:
            idf_scores[term] = math.log(total_docs / doc_freq)
        else:
            idf_scores[term] = 0.0

    return idf_scores

# Compute IDF for all terms using the positional index
idf_scores = compute_idf_from_positional_index(positional_index, len(documents))

# Example: Show IDF for a sample term
# NOTE : You could apply stemming in the normalization step, so the term may not match exactly if it was stemmed
sample_term = 'softwar' 
sample_term2 = 'applic'
print(f"IDF for term '{sample_term}': {idf_scores.get(sample_term, 0.0)}")
print(f"IDF for term '{sample_term2}': {idf_scores.get(sample_term2, 0.0)}")
```

### 🗣 Talking Point: How to Interpret IDF 

**What this code does**
- It computes **IDF** for each term in the positional index.
- For each term, it counts how many documents contain it: `doc_freq = len(doc_positions)`.
- It applies the formula: `idf = log(total_docs / doc_freq)`.
- Result: very common terms get low IDF; rare terms get high IDF.

**IDF value range in this implementation**
Using `IDF(t) = log(N / n_t)`, where:
- `N` = total number of documents.
- `n_t` = number of documents that contain term `t`.

For terms that appear in the corpus, IDF is in:
$$
0 \leq IDF(t) \leq \log(N)
$$

**Minimum value (0)**
- Happens when `n_t = N` (the term appears in all documents).
- Meaning: the term is not very discriminative for search (too generic).

**Maximum value (`log(N)`)**
- Happens when `n_t = 1` (the term appears in only one document).
- Meaning: the term is highly discriminative (more useful to separate documents).

**Practical note**
- If a term does not exist in the index, it is not included in the computation loop.
- In the example print, `idf_scores.get(sample_term, 0.0)` returns `0.0` when the term is not found.

### 📐 Mathematical Explanation: Inverse Document Frequency (IDF)

The **Inverse Document Frequency (IDF)** is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The intuition is that terms that appear in many documents are less informative than those that appear in few.

The IDF for a term $t$ is defined as:

$$
\text{IDF}(t) = \log\left(\frac{N}{n_t}\right)
$$

Where:
- $N$ is the total number of documents in the corpus.
- $n_t$ is the number of documents containing the term $t$.

---

#### Example Calculation

Suppose:
- $N = 20$ (total documents)
- $n_{\text{softwar}} = 15$ (documents containing the term 'softwar')

Then:

$$
\text{IDF}(\text{softwar}) = \log\left(\frac{20}{15}\right) \approx 0.2877
$$

So, the IDF for term 'softwar' is $0.2877$.

A higher IDF value means the term is rare across the corpus, while a lower value means the term is common. IDF is often used in combination with Term Frequency (TF) to compute the TF-IDF score:

$$
\text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t)
$$

Where $\text{TF}(t, d)$ is the frequency of term $t$ in document $d$.

This weighting helps highlight terms that are both frequent in a document and rare across the corpus, improving search relevance and document ranking.
